```python
import torch
import torch.nn as nn
```

```python
#creating our model using NN module
class Model(nn.Module):

  #creating the constructor
  def __init__(self,num_features):
    super().__init__()
    #adding layers
    self.layer1 = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()


  def forward(self,features):
    output = self.layer1(features)
    output = self.sigmoid(output)
    return output
```

```python
#ceating features
features = torch.rand(10,5) #10 rows 5 columns
```

```python
model11 = Model(features.shape[1])
```
