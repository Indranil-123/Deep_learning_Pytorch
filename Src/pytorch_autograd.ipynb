```python
#finding out the derivation using Autograd
x = torch.tensor(3.0,requires_grad=True)

#build the relation using function
y = x ** 2
```

```python
#extracting the derivation value
x.grad
```

```python
#now we are asigning
z = torch.sin(y)
```
